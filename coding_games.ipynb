```python
#!/usr/bin/env python3
"""
Rock - Paper - Scissors - Interactive Notebook Version

Ein interaktives Spiel (Schere, Stein, Papier) um Ihnen ein bisschen Python zu
zeigen und beizubringen.
"""

# Modul importieren
import random

#  Variablen anlegen
ROCK = "r"
PAPER = "p"
SCISSORS = "s"

TIE = "Unentschieden."
WIN = "Du gewinnst :-)"
LOSE = "Du verlierst :-("

# Funktion zum Vergleichen der Züge
def compare_moves(you, computer):
    """
    Vergleicht die Züge des Spielers und des Computers und gibt das Ergebnis zurück.
    """
    # Wenn--Dann-Vergleiche
    if you == computer:
        return TIE
    elif you == SCISSORS and computer == PAPER:
        return WIN
    elif you == ROCK and computer == SCISSORS:
        return WIN
    elif you == PAPER and computer == ROCK:
        return WIN
    else:
        # ... in allen anderen Fällen
        return LOSE

# Spiel starten
def play_game():
    """
    Startet das Spiel und erlaubt dem Spieler, mehrere Runden zu spielen.
    """
    while True:
        you = ""
        
        # While-Schleife für Benutzereingabe
        while you not in [ROCK, PAPER, SCISSORS]:
            # Eingabe
            you = input("(r)ock, (p)aper, or (s)cissors? > ").lower()
        
        # Funktionsaufrufe
        computer = random.choice([ROCK, PAPER, SCISSORS])
        result = compare_moves(you, computer)
        
        # Ausgabe
        print(f"Du: {you}, Computer: {computer}. {result}")
        
        # Wiederholung des Spiels
        play_again = input("Möchtest du noch eine Runde spielen? (ja/nein) > ").lower()
        if play_again != "ja":
            print("Danke fürs Spielen!")
            break

# Spielanleitung
print("Willkommen beim Rock-Paper-Scissors Spiel!")
print("Hier sind die Regeln:")
print("- Wähle einen der Züge: (r)ock, (p)aper, oder (s)cissors")
print("- Der Computer wählt zufällig einen Zug")
print("- Die Regeln sind einfach: Schere schlägt Papier, Papier schlägt Stein, Stein schlägt Schere")
print("- Viel Spaß!")

# Spiel starten
play_game()
```

```python
#!/usr/bin/env python3
"""
Guess the Number - Interactive Notebook Version

Ein interaktives Spiel, bei dem der Benutzer eine Zahl raten muss. Dieses Spiel
hilft Anfängern, grundlegende Programmierkonzepte zu verstehen.
"""

# Modul importieren
import random

# Spielanleitung
print("Willkommen beim 'Guess the Number' Spiel!")
print("Hier sind die Regeln:")
print("- Der Computer wählt zufällig eine Zahl zwischen 1 und 100.")
print("- Du musst diese Zahl erraten.")
print("- Nach jedem Versuch wird dir gesagt, ob deine Zahl zu hoch oder zu niedrig war.")
print("- Viel Spaß!")

# Spiel starten
def play_game():
    """
    Startet das Spiel und erlaubt dem Spieler, mehrere Runden zu spielen.
    """
    while True:
        # Der Computer wählt eine zufällige Zahl zwischen 1 und 100
        number_to_guess = random.randint(1, 100)
        guess = None
        attempts = 0

        # While-Schleife für die Benutzer, um die Zahl zu erraten
        while guess != number_to_guess:
            # Eingabe
            guess = input("Rate eine Zahl zwischen 1 und 100: ")
            
            # Überprüfen, ob die Eingabe eine Zahl ist
            if guess.isdigit():
                guess = int(guess)
                attempts += 1

                # Wenn--Dann-Vergleiche
                if guess < number_to_guess:
                    print("Zu niedrig!")
                elif guess > number_to_guess:
                    print("Zu hoch!")
                else:
                    print(f"Herzlichen Glückwunsch! Du hast die Zahl {number_to_guess} in {attempts} Versuchen erraten.")
            else:
                print("Bitte gib eine gültige Zahl ein.")
        
        # Wiederholung des Spiels
        play_again = input("Möchtest du noch eine Runde spielen? (ja/nein) > ").lower()
        if play_again != "ja":
            print("Danke fürs Spielen!")
            break

# Spiel starten
play_game()
```
